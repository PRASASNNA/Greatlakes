# 3.  Display the percentage distribution of the data in each year using the pie chart.

```python
year_counts = x['year'].value_counts()

# Get the labels and counts for the pie chart
labels = year_counts.index
counts = year_counts.values

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=80)

# Set a title
plt.title('Percentage Distribution of Data by Year')

# Display the pie chart
plt.show()
```

# 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python
plt.figure(figsize=(4,4))

sns.displot(data=x['priceperlb'], kde=True,bins=6,color='green',common_norm = True)

plt.title("Distribution Plot for price per lb ")
plt.show()
```

# 5.Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and wite your findings

```python

# Create the scatterplot
plt.scatter(x['numcol'], x['prodvalue'])

# Add labels and title
plt.xlabel('numcol')
plt.ylabel('prodval')
plt.title('Scatterplot of numcol vs. prodvale')
plt.grid(True)

# Show the plot
plt.show()
```

# 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
# Create a box plot of prodval by year
sns.boxplot(
    x = 'year',
    y = 'prodvalue',
    showmeans=True,
    data=x
)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Production value (USD)')
plt.title('Box plot of production value by year')
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```

# 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
# Assuming you have a DataFrame 'df' with the specified columns
selected_columns = ['numcol', 'yieldpercol', 'totalprod', 'prodvalue','year']

# Filter the DataFrame to include only the selected columns
df_selected = x[selected_columns]

# Create a pairplot
sns.set(style="ticks")
sns.pairplot(df_selected, hue="year")
plt.show()
```

# 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yieldpercol','totalprod', 'stocks', 'priceperlb', 'prodvalue')

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the correlation matrix
corr_matrix = x[['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']].corr()

# Display the correlation matrix as a heatmap
sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation matrix of honey production data')
plt.show()
```

Inferences
The heatmap shows that there is a strong positive correlation between the following variables:
1.numcol and prodval
2.numcol and totalprod
3.totalprod and prodval
This means that hives with more colonies tend to produce more honey and have a higher value.

There is also a moderate positive correlation between the following variables:

1.yieldpercol and prodval
2.yieldpercol and totalprod
3.This means that hives with a higher yield per colony tend to produce more honey and have a higher value.

However, there is a weak positive correlation between the following variables:

1.priceperlb and prodval
2.priceperlb and totalprod
This means that the price of honey per pound does not have a strong relationship with the production of honey.
