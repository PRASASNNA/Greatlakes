# 6. Check if there are missing values present in the column Rating, If any? drop them and and create a new column as 'Rating_category' by converting ratings to high and low categories(>3.5 is high rest low)


```python
#CHECK FOR MISSING  VALUES IN THE 'Rating' COLUMN AND DROP ROWS WITH MISSING VALUES
missing_values=data['Rating'].isna().sum()
if missing_values > 0:
    print(f"There are {missing_values} missing values in the column Rating.")
else:
    print("There are no missing values in the column Rating.")
```

```python
#DROP THE ROWS WITH MISSING VALUES IN THE COLUMN RATING
data=data.dropna(subset=['Rating'])
print("Missing values dropped.")
data
```

```python
#CREATE A NEW COLUMN AS "Rating_category" BY CONVERTING RATINGS TO HIGH AND LOW CATEGORIES

#DEFINE A FUNCTION TO ASSIGN CATEGORIES BASED ON RATINGS
def rating_category(rating):
    if rating > 3.5:
        return 'High'
    else:
        return 'Low'
    
#APPLY THE FUNCION TO THE COLUMN RATING AND CREATE A NEW COLUMN Rating_category
data['Rating_category']=data['Rating'].apply(rating_category)
data['Rating_category']
```

# 9. The column 'Size' contains alphanumeric values, treat the non numeric data and convert the column into suitable data type. (hint: Replace M with 1 million and K with 1 thousand, and drop the entries where size='Varies with device')

```python
#DEFINE A FUNCTION TO CONVERT SIZE TO NUMMERIC
def convert_size(size):
    if 'M' in size:
        size = size.replace('M', '')
        return float(size) * 1e6  # Convert to bytes (1M = 1e6 bytes)
    elif 'K' in size:
        size = size.replace('K', '')
        return float(size) * 1e3  # Convert to bytes (1K = 1e3 bytes)
    else:
        return np.nan  # Return NaN for 'Varies with device'
#APPLY THE FUNCTION TO THE 'Size'  column
data['Size'] = data['Size'].apply(lambda x: convert_size(x) if isinstance(x, str) else np.nan)
data['Size']
```

```python
data
```

```python
# Drop the entries where size is NaN (i.e., where original size was 'Varies with device')
data=data.dropna(subset=['Size'])
data
```

# 10. Check the column 'Installs', treat the unwanted characters and convert the column into a suitable data type.


```python
# Function to convert installs to numeric
def convert_installs(installs):
    installs = str(installs).replace(',', '')  # Remove commas
    installs = installs.replace('+', '')  # Remove plus sign
    return int(installs)  # Convert to integer

# Apply the function to the 'Installs' column
data['Installs'] = data['Installs'].apply(convert_installs)
data['Installs']
```
